```python
import pymongo
from pprint import pprint
from pymongo import MongoClient
client = MongoClient("localhost", 27017) #On se connecte a MongoDB
```

```python
#Affiche la liste des bases de données
print("\nListe des base de données")
for name in client.list_database_names():
    print(name)
```

```python
#Permet de rentrer dans la base de donnée gym ou la crée si elle n'existe pas
gym = client["gym"]
```

```python
#Permet de recuperer les collections de la database gym
Gymnases = gym.get_collection("Gymnases")
Sportifs = gym.get_collection("Sportifs")
```

```python
#Affiche les collections de la base gym
for name in gym.list_collection_names():
    print(name)
```

```python
print("\nAffiche tous les documents de la collection 'Gymnases':")
for y in Gymnases.find():
    pprint(y)
```

```python
print("\nAffiche tous les documents de la collection 'Sportifs':")
for y in Sportifs.find():
    pprint(y)
```

```python
print("\nAffiche le (ou les) conseiller(s) s’appelant 'KERVADEC:'")
for y in Sportifs.find({"Nom": "KERVADEC"}):
    pprint(y)
```

```python
print("\nAffiche les sportifs qui ont ou bien 32 ans ou 40 ans")
for y in Sportifs.find({"$or": [{"Age": 32}, {"Age": 40}]}):
    pprint(y)
```

```python
print("\nAffiche les sportifs jouant au Basket-ball ?")
for y in Sportifs.find({"Sports.Jouer": "Basket ball"}):
    pprint(y)
```

```python
print("\nAffiche les sportifs ayant au moins 32 ans, ou de sexe féminin ?")
for y in Sportifs.find({"$or": [{"Age": {"$gte": 32}}, {"Sexe": "F"}]}):
    pprint(y)
```

```python
#count_documents({}) permet de compter sur pymongo
print("\nAffiche le nombre de sportifs de sexe féminin dans la base.")
Comptage = Sportifs.count_documents({"Sexe": "F"})
pprint(Comptage)
```

```python
print("\nAffiche les sportifs (identifiant, nom et prénom) qui ont entre 20 et 30 ans ")
for y in Sportifs.find({"Age": {"$gte": 20,"$lte": 30}},{"Nom": 1, "Prenom" : 1,"Age":1}):
    pprint(y)
```

```python
print("\nAffiche les sportifs (identifiant, nom et prénom) qui pratiquent du hand ball ")
for y in Sportifs.find({},{"Nom": 1, "Prenom" : 1,"Sportifs.Jouer":"Hand ball"}:
    pprint(y)
```

```python
print("\nAffiche les gymnases de “Villetaneuse” ou de “Sarcelles” qui ont une surface de plus de 400 m2")
for y in Gymnases.find({"$or":[{"Ville":"VILLETANEUSE"},{"Ville":"SARCELLES"}],"Surface": 400}):
    pprint(y)
```

```python
print("\nAffiche les gymnases oû peut-on jouer au hockey le mercredi après 15H ")
for y in Gymnases.find({"Seances.Horaire":{"$gte": 15.0},"Seances.Libelle":"Hockey","Seances.Jour":"mercredi"},{"NomGymnase":1}):
    pprint(y)
```

```python
print("\nAffiche les gymnases qui n’ont pas de séances le dimanche.")
for y in Gymnases.find({"Seances.Jour":{"$nin":["Dimanche","dimanche"]}}):
    pprint(y)
```

```python
print("\nAffiche les gymnases qui ne proposent que des séances de basket ball ou de volley ball ")
for y in Gymnases.find({"$or":[{"Seance.Libelle":"Basket Ball"},{"Seance.Libelle":"Volley Ball"}]},{"Seance":{"$elemMatch":{"Libelle":"Basket Ball"}},"NomGymnase":1}):
    pprint(y)
```

```python
print("\nAffiche les gymnases qui ne proposent que des séances de basket ball ou de volley ball ")
for y in Gymnases.find({"$nor":[{"Seances.Libelle":{"$ne":"Basket ball"}},{"Seances.Libelle":{"$ne":"Volley ball"}}]}):
    pprint(y)
```

```python
print("\nAffiche les entraineurs qui sont toujours joueurs")
for y in Sportifs.find({"$and":[{"Sports.Entrainer":{"$exists":True}},{"Sports.Jouer":{"$exists":True}}]}):
    pprint(y)
```

```python
print("\nAffiche le nom du conseiller du sportif 'Kervadec' ")
for y in Sportifs.find({"IdSportif":1},{"Nom":1}):
    pprint(y)
```

```python
print("\nAffiche les entraineurs qui entrainent du hand et du basket.")
for y in Sportifs.find({"$and":[{"Sports.Entrainer":"Basket ball"},{"Sports.Entrainer":"Hand ball"}]},{"Nom":1}):
    pprint(y)
```

```python
print("\nAffiche les sportifs qui n'ont pas de conseiller.")
for y in Sportifs.find({"IdSportifConseiller":{"$exists":False}},{"Nom":1}):
    pprint(y)
```
